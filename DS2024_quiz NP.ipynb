პასუხი: ეს ისტორიები არის ნათელი მაგალითი იმისა, თუ როგორ უნდა შევაგროვოთ და დავამუშაოთ მონაცემები, რომ შევქმათ პროგრამა, რომელიც ამ მონაცემებზე დაყრდნობით შეძლებს ამოარჩიოს საუკეთესო ვარიანტი სიტუაციის მიხედვით და გააკეთოს საუკეთესო სვლა სტატისტიკაზე და მიწოდებულ ლოგიკაზე დაყრდნობით(ჭადრაკისა და GO შემთხვევაში). მონაცემთა მეცნიერების წარმატების ისტორიად შეგვიძია ჩავთვალოთ ყველა ის ინციდენტი, სადაც მონაცემების შეგროვებითა და დამუშავებით შეგვიძია შევქმნათ მსგავსი ჭკვიანი პროგრამები, რომლებიც სწორად ამუშავებენ და აანალიზებენ დიდი რაოდენობის მონაცემებს და ამ მონაცემთა ანალიზის შედეგად აკეთებენ პროგნოზებს, რომლებიც მართლდება და რეალურ შედეგს გვაძლევს.

პასუხი: ყოველდღიურად ადამიანები ვიღებთ და ვამუშავებთ მონაცემებს. გამოცდილებიდან მქონია შემთხვევები, როდესაც მონაცემების სწორი ანალიზითა და დაკვირვებით უკეთესი შედეგი მიმიღია. მოვიყვან მაგალითს: სკოლის პერიოდში ვიყენებდი სქემას, რომელიც ახლო კავშირშია ამ დავალებასთან. სქემის მიხედვით მონაცემებზე დაყრდნობით ვცდილობდი გამეგო როგორი რეჟიმი იყო ჩემთვის ყველაზე ეფექტური მეცადინეობის კუთხით. სკოლიდან მოსვლის შემდეგ ვინიშნავდი დროს და ვიწყებდი მეცადინეობას. მქონდა რამდენიმე ვარიანტი დასვენებისა და სამუშაო დროის განაწილების და შემდეგ ვაფასებდი, რომელი განაწილება უფრო ეფექტური იყო (ნაკლებ დროში მეტი ცოდნის მიღება). საბოლოო ჯამში მივედი იმ დასკვნამდე, რომ ჩემთვის ყველაზე ეფექტური იყო 45 წუთი მუშაობა და შემდეგ 15 წუთი დასვენება. დანარჩენი ორი ვარიანტი ნაკლებ ეფექტური აღმოჩნდა, საცდელად მქონდა აღებული 1 საათი მუშაობა 15 წუთი დასვენება და ნახევარი საათი მუშაობა 10 წუთი დასვენება. წლების განმავლობაში ამ სქემას ვცვლიდი ხოლმე და ახალ ვარიანტებს ვცდილობდი. შემდეგ დაგროვილ ინფორმაციას თუ რისი გაკეთება მოხერდა რა დროში ვაანალიზებდი და გამომქონდა დასკვნა თუ რომელი ვარინტი უფრო მისაღები იყო ჩემთვის.
